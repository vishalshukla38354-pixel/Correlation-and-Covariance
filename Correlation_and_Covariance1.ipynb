Q1. Define Covariance and explain how it differs from Correlation in terms of scale and interpretation.

Ans:

**Covariance** measures how two variables change together.

- if both variables increase or decrease together, covareance is positive.
- if one increases while the other decreases, covariance is negative.


**Correlation** measures both the **strength and direction** of the relationship between two variables.

**Difference:**

Covariance:
-
- Depends  on scale/units
- Value can be any number
- Hard to interpret


Correlation:
-
- Scale - free
- Value lies between -1 and +1
- Easy to intterpret

Q2. What does a positive, negative, and zero covariance indicate about the relationship between two variables?

Ans:


- **Positive convariance:** Both variables move in the same direction.
- **Negative  convariance:** Variables move in opposite direction.
- **Zero covariance:**  No linear relationship between variables.

Q4. Explain the difference between Pearson’s correlation coefficient and Spearman’s rank correlation
coefficient. When would you prefer to use Spearman’s correlation?

Ans:

Pearson Correlation:
-
- Measures lineear relationship
- Uses actual values
- Sensitive to outliers

Spearman Rank Correlation
-
- Measures monotonic relationship
- uses ranks
- Less sensitive to outliers

Use **Spearman correlation** when

- Data is ranked
- Data is not normally distributed
- Relationship is not linear.

Q5. If the correlation coefficient between two variables X and Y is 0.85, interpret this value in context. Can you
infer causation from this value? Why or why not?


Ans:

A correlation coefficient of 0.85 indicates a **strong positive relationship** between X and Y.

This means:

- As X increases, Y also tends to increase.

But **correlation does not imply causation.**

We cannot say X causes Y because other factors may influence the relationship.
